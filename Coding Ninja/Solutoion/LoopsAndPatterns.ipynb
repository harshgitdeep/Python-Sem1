```python
# Program to check the Happy number
# Its a number which is replaced by the sum of square of its digits till the sum converges to 1
```

```python
sum=None
for n in range(3,20):
    sum=0
    print("Number: ",n,"=>",end=" ")
    while True:
        sum=0
        while n!=0:
            rem=n%10
            n=n//10   # 1234//10  => 123
            sum+=rem * rem
        print(sum,end=" ")
        n=sum
        if sum==1 or sum==4:
            break
    if sum==1:
        print("Its Happy Number")
    elif sum==4:
        print("Not A HAPPY Number")
    
```

```python
# Searching Prime Number
for number in range(2,10):
    count=0
    for divisor in range(1,number+1):
        if number%divisor==0:
            count+=1
    print("Number: ",number," No of Divisors :",count,end=" ")  
    if count==2:
        print("PRIME")
    else:
        print()
```

```python
for number in range(2,20):
    for divisor in range(2,number//2+1):
        if number%divisor==0:
            print(number, "is Not a Prime",divisor)
            break
    else:
        print(number," is Prime")
```

```python
# a Number should be divided by the MINIMUM number which is less than or equal to its sqrare root value
import math
for number in range(2,30):
    for divisor in range(2,int(math.sqrt(number))+1):
        if number%divisor==0:
            print(number, "is Not a Prime",divisor)
            break;
    else:
        print(number," is Prime")
```
